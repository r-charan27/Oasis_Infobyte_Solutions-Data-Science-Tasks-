```python
import pandas as pd
# Load dataset
df = pd.read_csv("C:/Users/RAMCHARAN/Downloads/spam.csv", encoding='ISO-8859-1')
# Preview
print(df.head())
print(df.columns)
```

```python
# Keep only the needed columns (usually first two)
df = df[['v1', 'v2']]
df.columns = ['label', 'message']

# Map labels to binary values
df['label'] = df['label'].map({'ham': 0, 'spam': 1})
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.model_selection import train_test_split
```

```python
# Split data
X_train, X_test, y_train, y_test = train_test_split(df['message'], df['label'], test_size=0.2, random_state=42)
```

```python
# Vectorization
vectorizer = TfidfVectorizer(stop_words='english')
X_train_vec = vectorizer.fit_transform(X_train)
X_test_vec = vectorizer.transform(X_test)
```

```python
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

# Train
model = MultinomialNB()
model.fit(X_train_vec, y_train)
```

```python
# Predict
y_pred = model.predict(X_test_vec)
# Evaluate
print("Accuracy:", accuracy_score(y_test, y_pred))
print("\nClassification Report:\n", classification_report(y_test, y_pred))
```

```python
test_msg = ["Congratulations! You've won a $1000 gift card. Click here to claim."]
test_vec = vectorizer.transform(test_msg)
print("Spam" if model.predict(test_vec)[0] else "Not Spam")
```

```python
# Examples to classify
examples = [
    "Congratulations! You've won a $1000 gift card. Click here to claim.",  # Spam
    "Hey, are we still meeting for lunch at 1 PM today?"                     # Not Spam
]

# Transform the examples using the same vectorizer
example_vectors = vectorizer.transform(examples)

# Predict
predictions = model.predict(example_vectors)

# Output results
for msg, pred in zip(examples, predictions):
    print(f"Message: \"{msg}\"\nPrediction: {'Spam' if pred == 1 else 'Not Spam'}\n")
```
