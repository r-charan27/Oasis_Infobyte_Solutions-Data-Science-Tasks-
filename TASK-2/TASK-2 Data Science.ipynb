```python
import pandas as pd 
df=pd.read_csv("C:/Users/RAMCHARAN/Downloads/Unemployment in India.csv")
print(df.head())
```

```python
print(df.info())
print(df.describe())
print(df.columns)
```

```python
# Fix column names
df.columns = df.columns.str.strip()

# Check column names
print(df.columns.tolist())

# Convert 'Date' to datetime
df['Date'] = pd.to_datetime(df['Date'])

# Display a few rows
print(df.head())
```

```python
#Cleaning and Preparing the Data
df['Date'] = pd.to_datetime(df['Date'])
df = df.rename(columns=lambda x: x.strip()) 
```

```python
import matplotlib.pyplot as plt

df.groupby('Date')['Estimated Unemployment Rate (%)'].mean().plot(figsize=(10, 5))
plt.title("Average Unemployment Rate in India Over Time")
plt.ylabel("Unemployment Rate (%)")
plt.xlabel("Date")
plt.grid()
plt.show()
```

```python
import seaborn as sns

plt.figure(figsize=(12, 6))
sns.boxplot(data=df, x='Region', y='Estimated Unemployment Rate (%)')
plt.xticks(rotation=90)
plt.title("Unemployment Rate Distribution by Region")
plt.show()
```

```python
covid_df = df[df['Date'].dt.year == 2020]
covid_df.groupby('Date')['Estimated Unemployment Rate (%)'].mean().plot(figsize=(10, 5))
plt.title("Unemployment Rate During COVID-19 (2020)")
plt.ylabel("Unemployment Rate (%)")
plt.xlabel("Date")
plt.grid()
plt.show()
```
